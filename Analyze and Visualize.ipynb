# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = np.zeros(10)
arr[4] = 1
arr_2d = arr.reshape(2, 5)
print(arr_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)
arr_reversed = arr[::-1]
even_elements = arr_reversed[arr_reversed % 2 == 0]
print("Развернутый массив:", arr_reversed)
print("Четные элементы:", even_elements)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr_2d = np.arange(9).reshape(3, 3)
print(arr_2d)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.random((4, 3, 2))
min_value = arr.min()
max_value = arr.max()
print("Массив:\n", arr)
print("Минимальное значение:", min_value)
print("Максимальное значение:", max_value)
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr1 = np.random.randint(10, size=(6, 4))
arr2 = np.random.randint(10, size=(4, 3))
result = np.matmul(arr1, arr2)
print("Массив 6x4:\n", arr1)
print("Массив 4x3:\n", arr2)
print("Результат матричного умножения (6x3):\n", result)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.random((7, 7))
mean = arr.mean()
std_dev = arr.std()
normalized_arr = (arr - mean) / std_dev
print("Исходный массив:\n", arr)
print("Среднее значение:", mean)
print("Стандартное отклонение:", std_dev)
print("Нормализованный массив:\n", normalized_arr)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
tips.head(5)
```

### Узнайте сколько всего строчек и колонок в данных

```python
rows, columns = tips.shape
print(f"Количество строк: {rows}")
print(f"Количество колонок: {columns}")
```

### Проверьте есть ли пропуски в данных

```python
missing_values = tips.isnull().sum()
print("Количество пропусков в каждом столбце:\n", missing_values)
```

### Посмотрите на распределение числовых признаков

```python
numeric_summary = tips.describe()
print("Распределение числовых признаков:\n", numeric_summary)
```

### Найдите максимальное значение 'total_bill'

```python
max_total_bill = tips['total_bill'].max()
print("Максимальное значение 'total_bill':", max_total_bill)
```

### Найдите количество курящих людей

```python
num_smokers = tips[tips['smoker'] == 'Yes'].shape[0]
print("Количество курящих людей:", num_smokers)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
average_total_bill_by_day = tips.groupby('day', observed=False)['total_bill'].mean()
print("Средний 'total_bill' в зависимости от 'day':\n", average_total_bill_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_total_bill = tips['total_bill'].median()
filtered_tips = tips[tips['total_bill'] > median_total_bill]
average_tip_by_sex = filtered_tips.groupby('sex', observed=False)['tip'].mean()
print("Средний 'tip' в зависимости от 'sex' для строк с 'total_bill' больше медианы:\n", average_tip_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1})
print("Первые строки с преобразованным признаком 'smoker':\n", tips.head())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Распределение 'total_bill'")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Взаимосвязь между 'total_bill' и 'tip'")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Взаимосвязь между 'total_bill' и 'day'")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=tips, x='tip', hue='time', bins=20, kde=True, element="step")
plt.title("Распределение 'tip' в зависимости от категорий 'time'")
plt.xlabel("Tip")
plt.ylabel("Frequency")
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=tips[tips['sex'] == 'Male'], x='total_bill', y='tip', hue='smoker', palette='coolwarm')
plt.title("Взаимосвязь 'total_bill' и 'tip' для Male")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.subplot(1, 2, 2)
sns.scatterplot(data=tips[tips['sex'] == 'Female'], x='total_bill', y='tip', hue='smoker', palette='coolwarm')
plt.title("Взаимосвязь 'total_bill' и 'tip' для Female")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.tight_layout()
plt.show()
```
